## Data cleaning (Rank Column)
1. Remove  the string attributes from the column
2. Convert the column into an integer

```python
data.head()
```

```python
data['Rank'] = data['Rank'].str[0:-2]
```

```python
data['Rank'] = data['Rank'].str.replace(',', '')
```

```python
data.tail()
```

```python
data['Rank'] = data['Rank'].astype(int)
```

```python
data.dtypes
```

## Data cleaning (Video uploads and subscribers)

```python
data.head()
```

```python
data.dtypes
```

```python
data['Video Uploads']= data['Video Uploads'].astype(int)
```

```python
data['Subscribers'] = data['Subscribers'].astype(int)
```

```python
data.dtypes
```

## Data cleaning (Grade column)

```python
data.head()
```

```python
data['Grade'].unique()
```

```python
data['Grade'] = data['Grade'].map({'A++ ': 5, 'A+ ': 4, 'A ': 3, 'A- ': 2, 'B+ ': 1})
```
